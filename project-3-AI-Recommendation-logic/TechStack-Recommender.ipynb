```python
import pandas as pd 
df = pd.read_csv("data/raw_skills.csv")
print(df.head())
```

```python
print(df)
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
skill_matrix = tfidf.fit_transform(df["Skills"])
```

```python
print(skill_matrix.shape)
```

```python
user_input = input("Enter your skills (separated by commas):")
```

```python
user_text = user_input.replace(",", " ")
```

```python
user_vector = tfidf.transform([user_text])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(user_vector, skill_matrix)
```

```python
print(similarity_scores)
```

```python
df["Similarity"] = similarity_scores.flatten()
```

```python
print(df.columns.tolist())
```

```python
df[["Job_Role", "Similarity"]]
```

```python
df = df.sort_values(by="Similarity", ascending=False)
```

```python
top3 = df.head(3)
print(top3)
```

```python
print(top3[["Job_Role", "Similarity"]])
```

```python
df["Match_Percentage"] = (df["Similarity"] * 100).round(2)
```

```python
print(df[["Job_Role", "Match_Percentage"]].head(3))
```

```python
print("====== Top 3 career Recommendations =====")

for index, row in df.head(3).iterrows():
    print(f"{row['Job_Role']} --> {row['Match_Percentage']}% Match")
```
